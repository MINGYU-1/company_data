```python
import mysql.connector
import os
from dotenv import load_dotenv
load_dotenv()

db_password = os.getenv('password')

conn = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password=db_password,
    database="new_schema"
)
```

```python
cur = conn.cursor()
cur.execute("SHOW TABLES")
```
