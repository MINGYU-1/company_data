```python
import pandas as pd
from pyproj import Transformer

# === 설정 ===
csv_path = "../data_csv/11_09/강원도_사업체_공간집계정보.csv"
output_csv = "강원도_사업체_위경도_변환.csv"

# EPSG:5179 → EPSG:4326 (WGS84 위도/경도)
transformer = Transformer.from_crs("EPSG:5179", "EPSG:4326", always_xy=True)

# CSV 로드
try:
    df = pd.read_csv(csv_path, encoding="utf-8")
except:
    df = pd.read_csv(csv_path, encoding="cp949")

print("원본 컬럼:", df.columns)

# 변환 함수
def xy_to_latlon(x, y):
    lon, lat = transformer.transform(x, y)
    return lat, lon

# 새 컬럼 추가
df["UPLFT_LAT"], df["UPLFT_LON"] = zip(*df.apply(lambda r: xy_to_latlon(r["UPLFT_XCNTS"], r["UPLFT_YDNTS"]), axis=1))
df["LOWRGT_LAT"], df["LOWRGT_LON"] = zip(*df.apply(lambda r: xy_to_latlon(r["LOWRGT_XCNTS"], r["LOWRGT_YDNTS"]), axis=1))

# 저장
df.to_csv(output_csv, index=False, encoding="utf-8")
print("완료:", output_csv)
```

```python
import pandas as pd

# === 설정 ===
csv_path = "../data_csv/11_09/강원도_사업체_공간집계정보.csv"   # ← CSV 경로만 바꾸면 됨
sql_output = "insert_kangwondo_Spatial_Statistics.sql"
table_name = "spatial_statistics"
# ==============

def clean(v):
    """MySQL용 값 변환"""
    if pd.isna(v):
        return "NULL"
    if isinstance(v, str):
        v = v.replace("\\", "\\\\").replace("'", "''")
        return f"'{v}'"
    return str(v)

# CSV 읽기 (인코딩 자동 처리)
try:
    df = pd.read_csv(csv_path, encoding="utf-8")
except UnicodeDecodeError:
    df = pd.read_csv(csv_path, encoding="cp949")

print("CSV 읽기 성공:", df.shape)
print("컬럼명:", list(df.columns))

# CREATE TABLE 자동생성
col_types = []
for col in df.columns:
    if df[col].dtype == "float64":
        col_types.append(f"{col} DOUBLE")
    elif df[col].dtype == "int64":
        col_types.append(f"{col} BIGINT")
    else:
        col_types.append(f"{col} VARCHAR(255)")

create_sql = f"DROP TABLE IF EXISTS {table_name};\n"
create_sql += f"CREATE TABLE {table_name} (\n"
create_sql += ",\n".join(["    " + c for c in col_types])
create_sql += "\n);\n\n"

# INSERT 생성
rows_sql = []
for row in df.itertuples(index=False, name=None):   # ← 여기 수정됨
    values = [clean(v) for v in row]                # ← 컬럼명 대신 튜플 값 그대로 사용
    rows_sql.append("(" + ", ".join(values) + ")")

insert_sql = (
    f"INSERT INTO {table_name} ({', '.join(df.columns)}) VALUES\n" +
    ",\n".join(rows_sql) +
    ";"
)

# SQL 파일 저장
with open(sql_output, "w", encoding="utf-8") as f:
    f.write("-- SQL generated automatically\n\n")
    f.write(create_sql)
    f.write(insert_sql)

print(f"SQL 파일 생성 완료: {sql_output}")
```
