```python
import pandas as pd
df = pd.read_csv('m5_state_sales.csv')
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
```

```python
df.info()
```

```python
df['state_id'].unique()
```

```python
import matplotlib.pyplot as plt
Target = 'sales'
# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 6)) # Adjust size as needed

# Plot each series on the same axes (ax=ax)
df[df['state_id'] == 'CA'][Target].plot(ax=ax, label='CA Sales')
df[df['state_id'] == 'TX'][Target].plot(ax=ax, label='TX Sales')
df[df['state_id'] == 'WI'][Target].plot(ax=ax, label='WI Sales')

# Add a title, labels, and the legend
ax.set_title(f'Sales Comparison by State')
ax.set_xlabel('Date/Index')
ax.set_ylabel(Target)
ax.legend()

# Display the plot
plt.show()
```
